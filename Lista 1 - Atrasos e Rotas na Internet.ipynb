## i. Atrasos observados e caracterizados dos enlaces envolvidos

Há quatro tipos de fonte de atrasos de pacotes:

    1 - Processamento nos nós
    2 - Enfileiramento
    3 - Atraso de transmissão
    4 - Atraso de progpagação

No Processamento nos nós o atraso é dado pois é verificado os erros de bit e o link de saída é determinado. 

No Enfileiramento, temos um atraso pois há o tempo de espera no link de saída para a transmissão e isso está diretamente relacionado ao nível de congestionamento do roteador (quanto mais congestioinado, mais atraso).

No Atraso de transmissão é relacionado a larguda de banda do link **R** em $[bps]$ com o tamanho do pacote **L** em $[bits]$, obtendo, assim, uma relação de tempo **t** para enviar bits ao link dada por:

$\qquad t = \frac{L}{R}$

No Atraso de propagação é relacionado o comprimento do link físico **d** com a velocidade **s** de propagação no meio (dada por $\sim 2x10^8 m/s$), obtendo, assim a seguinte relação:

$\qquad A_{prop} = \frac{d}{s}$

Analisando a saída do comando `tracert www.caltech.edu`, observamos que no sexto salto temos, provavelmente, uma conexão com um roteador fora da cidade de Campinas, pois há um leve aumento nos tempos (tempos menores que $10ms$ para tempos maiores que $10ms$). Vemos, também esta relação de tempo maiores nos saltos subsequêntes.

# Discussão e Conclusão

Em nossa análise, nota-se uma discordância entre a distância SP/Brasil → CA/Estados Unidos, pois torna-se inverossímel termos praticamente $10.000km$ percorridos em apenas $10ms$ aproximadamente. Com isso em mente, fomos pesquisar e nos informar melhor da situação que nos encontramos com o endereço www.caltech.edu.

No seguinte link https://www.imperva.com/learn/performance/round-trip-time-rtt/ encontramos informações sobre como reduzir o *RTT* (*Round Trip Time*, em português: Tempo de Ida e Volta) de uma conexão: através de *CDN* (*Content Delivery Networks*, em português: Redes de Entrega de Conteúdo) onde, através deste link https://www.imperva.com/learn/performance/what-is-cdn-how-it-works/, nos diz que a missão do CDN é encurtar virtualmente a distância física entre origem e destino, com o objetivo de melhorar a velocidade e o desempenho da renderização de um site.

O modo de funcionamento do CDN é através de memórias em cache: focando na minimização da distância entre os visitantes e o servidor de um site, CDN armazena uma versão em cache do conteúdo do site em várias localizações geográficas (também conhecidas como Pontos de Presença ou *PoPs*, em inglês *Points of Presence*). Cada PoP contém vários servidores de cache responsáveis pela entrega de conteúdo aos visitantes em sua proximidade.

A partir disso, é possível entendermos a razão do tempo pequeno de $10ms$ para que fosse "percorrido" $10.000km$, o endereço www.caltech.edu estava na memória cache e, assim o conteúdo do endereço foi entregue rapidamente.

Observando ainda mais os resutlados das pesquisas dos IPs que obtivemos no `tracert`, temos a Cloudflare como organização responsável do link www.caltech.edu. Ao entrar no site da Cloudflare https://www.cloudflare.com/pt-br/network/, vemos que temos PoPs em Campinas e em São Paulo capital, explicando, mais uma vez, a razão de uma conexão tão rápida com e universidade Caltech.
